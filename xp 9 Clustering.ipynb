```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Prepare data
n_samples = 500
n_columns = 2
data = np.random.randn(n_samples, n_columns) * 2 + [5, 5]
```

```python
data = pd.DataFrame(data, columns=['Views', 'Time Spent'])
```

```python
# Plot data
plt.scatter(data['Views'], data['Time Spent'], color='gray', alpha=0.25)
plt.xlabel('Views')
plt.ylabel('Time Spent')
plt.title('Web visitor data')
plt.show()
```

```python
# Perform k-means clustering
n_clusters = 5
kmeans = KMeans(n_clusters = n_clusters, random_state=42)
data['Clusters'] = kmeans.fit_predict(data)
```

```python
# Visualize clusters
colors = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'magenta']

for cluster in range(n_clusters):
    cluster_data = data[data['Clusters'] == cluster]
    plt.scatter(cluster_data['Views'], cluster_data['Time Spent'], c=colors[cluster], label=f'Cluster {cluster + 1}', alpha=0.25)

plt.xlabel('Views')
plt.ylabel('Time Spent')
plt.title('Vistors data')
plt.show()
```
